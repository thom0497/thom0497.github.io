```python
from sklearn import tree
import pandas as pd
import os
```

```python
QB_data = pd.read_csv("../nfl-project3/QB Machine Learning.csv")
```

```python
QB_data = QB_data.loc[QB_data["Touches"] > 0]
del QB_data["VertLeap"]
del QB_data["BenchPress"]
del QB_data["ThreeCone"]
del QB_data["Shuttle"]
del QB_data["Arm Length"]
del QB_data["BroadJump"]
del QB_data["BroadJump"]
QB_data
```

```python
target = QB_data["RookieAV"]
```

```python
data = QB_data.drop("RookieAV", axis=1)
data.fillna(data.mean(), inplace=True)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
clf = tree.DecisionTreeRegressor()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=10000)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
row = [[74.125, 225, 9.25, 24, 4.8, 1657, 63.5, 8.3, .069, .021, .071, 2.7]]
yhat = rf.fit(X_train, y_train).predict(row)
print('Prediction: %d' % yhat[0])
```
