```python
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
```

```python
master_data = pd.read_csv("../data/Draft_SD_CSV.csv")
```

```python
#Grouping drafted player data to see production at franchise level

data_grouped_by_year_franch = master_data.groupby(["Year", "Current_Franchise"]).sum()
```

```python
data_grouped_by_year_franch
```

```python
#Creating csv for bar graph

data_grouped_by_year_franch.to_csv("../data/Data_Grouped_by_Year_and_Franchise.csv")
```

```python
#Gathering data and future Win Percentage by franchise.

win_data = pd.read_csv("../data/MLB_Draft_ThreeToFiveYearWPCT.csv")
```

```python
win_data
```

```python
#Merging Draft Dataset and Franchise Future Win Percentage

combined_data = pd.merge(data_grouped_by_year_franch, win_data, left_on=["Year", "Current_Franchise"], right_on=["yearID", "franchName"], how = "inner")
```

```python
combined_data
```

```python
#Examining correlation between win percentage and Team Control WAR

slope, y_int, r_value, p_value, std_err = st.linregress(combined_data["TC_Total_WAR"], combined_data["ThreeYearWPCT"])

fit = combined_data["TC_Total_WAR"] * slope  + y_int
equation = "y= " + str(round(slope,3)) + "x + " + str(round(y_int,3))
r_square = "R-Squared Value is " + str(round(r_value,2))
plt.scatter(combined_data["TC_Total_WAR"], combined_data["ThreeYearWPCT"], facecolor = "skyblue")
plt.plot(combined_data["TC_Total_WAR"], fit, color = "red")
plt.annotate(equation, (20, .45), color = "red", fontsize = 12)
plt.annotate(r_square, (19, .42), color = "red", fontsize = 12)
plt.title("Team Control WAR vs. Win Pct in Years 3-5 After the Draft")
plt.xlabel("Team Control WAR")
plt.ylabel("Drafting Team Win Pct")
plt.show()
```

```python
#Transforming the Team-Control WAR variable to attempt to get a more normal distribution

combined_data["log_value"] = np.log(combined_data["TC_Total_WAR"] + 10)
```

```python
#Examining correlation with transformed data

slope, y_int, r_value, p_value, std_err = st.linregress(combined_data["log_value"], combined_data["ThreeYearWPCT"])

fit = combined_data["log_value"] * slope  + y_int
equation = "y= " + str(round(slope,3)) + "x + " + str(round(y_int,3))
r_square = "R-Squared Value is " + str(round(r_value,2))
plt.scatter(combined_data["log_value"], combined_data["ThreeYearWPCT"], facecolor = "skyblue")
plt.plot(combined_data["log_value"], fit, color = "red")
plt.annotate(equation, (2.5, .45), color = "red", fontsize = 12)
plt.annotate(r_square, (2, .42), color = "red", fontsize = 12)
plt.title("Ln Team Control WAR vs. Win Pct in Years 3-5 After the Draft")
plt.xlabel("Ln Team Control WAR")
plt.ylabel("Drafting Team Win Pct")
plt.show()
```
