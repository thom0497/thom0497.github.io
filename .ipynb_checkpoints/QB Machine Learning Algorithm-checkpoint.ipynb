```python
#Dependencies

from sklearn import tree
import pandas as pd
import os
```

```python
#Importing QB Data

QB_data = pd.read_csv("../nfl-project3/QB Machine Learning.csv")
```

```python
#Filtering QB Data by Frequency

QB_data = QB_data.loc[QB_data["Touches"] > 100]

QB_data
```

```python
#Initialize Rookie AV

target = QB_data["RookieAV"]
```

```python
#Prepare QB data for modeling

data = QB_data.drop(["Year", "Name", "College", "POS", "SixtyYdShuttle", "BenchPress", "ThreeCone", "BroadJump", "VertLeap", "Shuttle", "RookieAV"], axis=1)
data.fillna(data.mean(), inplace=True)
feature_names = data.columns
data.head()
```

```python
#Randomly split the data into train and test data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=20)
```

```python
#Create one regression tree and find the score

clf = tree.DecisionTreeRegressor()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
#Create a random forest regression and find the score

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=10000)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
#Show which variables impact the model the most

sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
#Import 2017 QB data and fit it to the model

NewQBs = pd.read_csv("../nfl-project3/Quarterbacks 2017.csv")

NewQBs.fillna(data.mean(), inplace=True)

predictionDF = NewQBs.drop(["Year", "Name", "College", "POS"], axis = 1)

yhat = rf.fit(X_train, y_train).predict(predictionDF)
```

```python
NewQBs
```

```python
#Create a dataframe for the 2017 draft class results

InProgressQBs = pd.DataFrame({"Player" : NewQBs["Name"], "College" : NewQBs["College"], "Prediction" : yhat})
```

```python
#Sort the players by greatest value

InProgressQBs.sort_values(by="Prediction", ascending = False)
```
